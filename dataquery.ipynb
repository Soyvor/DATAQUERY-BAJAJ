```python
import pandas as pd
from datetime import datetime

# Define the calculate_age function
def calculate_age(birth_date_str):
    if pd.isna(birth_date_str) or not isinstance(birth_date_str, str):
        return None
    try:
        birth_date = datetime.strptime(birth_date_str, '%Y-%m-%d')
        today = datetime.today()
        age = today.year - birth_date.year - ((today.month, today.day) < (birth_date.month, birth_date.day))
        return age
    except ValueError:
        return None

# Define the categorize_age_group function
def categorize_age_group(age):
    if pd.isna(age):
        return 'Unknown'
    elif age <= 12:
        return 'Child'
    elif age <= 19:
        return 'Teen'
    elif age <= 59:
        return 'Adult'
    else:
        return 'Senior'

# Sample DataFrame
data = {
    'patientDetails_birthDate': ['2010-05-15', '2005-08-22', '1980-01-30', '1955-11-05', None]
}
df = pd.DataFrame(data)

# Calculate age
df['age'] = df['patientDetails_birthDate'].apply(calculate_age)

# Categorize age into age groups
df['ageGroup'] = df['age'].apply(categorize_age_group)

# Display the DataFrame
print(df)
```

```python
# prompt: What is count of Adult?

adult_count = df[df['ageGroup'] == 'Adult']['ageGroup'].count()
print("Count of Adults:", adult_count)
```
