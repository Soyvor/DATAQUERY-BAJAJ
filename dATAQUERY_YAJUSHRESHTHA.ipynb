```python
# prompt: Calculate the average number of medicines prescribed in the dataset.

import pandas as pd
import json
from google.colab import files
# Import necessary libraries

# Upload the JSON file
uploaded = files.upload()

# Load the JSON data
file_name = list(uploaded.keys())[0]
with open(file_name, 'r') as file:
    data = json.load(file)

# Normalize the JSON data into a flat table
df = pd.json_normalize(data, sep='_')

# Display the first few rows of the dataframe and its columns
print(df.head())
print(df.columns)


# Calculate missing values for each column
missing_values = df.isnull().sum()

# Print the missing values
print(missing_values)


# Calculate percentage of missing values for specific columns
missing_percentage_firstName = (df['patientDetails_firstName'].isnull().sum() / len(df)) * 100
missing_percentage_lastName = (df['patientDetails_lastName'].isnull().sum() / len(df)) * 100
missing_percentage_birthDate = (df['patientDetails_birthDate'].isnull().sum() / len(df)) * 100

# Print the percentages
print("Percentage of missing values in 'patientDetails_firstName': {:.2f}%".format(missing_percentage_firstName))
print("Percentage of missing values in 'patientDetails_lastName': {:.2f}%".format(missing_percentage_lastName))
print("Percentage of missing values in 'patientDetails_birthDate': {:.2f}%".format(missing_percentage_birthDate))


# Impute missing values in 'patientDetails_gender' with the mode
mode_gender = df['patientDetails_gender'].mode()[0]
df['patientDetails_gender'].fillna(mode_gender, inplace=True)

# Calculate the percentage of female gender after imputation
female_percentage = (df['patientDetails_gender'] == 'F').sum() / len(df) * 100

# Print the percentage
print("Percentage of female gender after imputation: {:.2f}%".format(female_percentage))

# Calculate the number of medicines prescribed for each patient
df['num_medicines'] = df['medications'].apply(lambda x: len(x) if isinstance(x, list) else 0)

# Calculate the average number of medicines prescribed
average_medicines = df['num_medicines'].mean()

# Print the average number of medicines
print("Average number of medicines prescribed: {:.2f}".format(average_medicines))
```

```python
# prompt: Add a column ageGroup that categorizes patients into age groups (e.g., Child for ages 0-12, Teen for ages 13-19, Adult for ages 20-59, Senior for ages 60+).

# Define a function to categorize age groups
def categorize_age(birth_date):
  # Assuming birth_date is in YYYY-MM-DD format
  from datetime import datetime
  today = datetime.today()
  birth_date = datetime.strptime(birth_date, '%Y-%m-%d')
  age = today.year - birth_date.year - ((today.month, today.day) < (birth_date.month, birth_date.day))
  if age <= 12:
    return 'Child'
  elif 13 <= age <= 19:
    return 'Teen'
  elif 20 <= age <= 59:
    return 'Adult'
  else:
    return 'Senior'

# Apply the function to create the 'ageGroup' column
df['ageGroup'] = df['patientDetails_birthDate'].apply(categorize_age)

# Display the updated dataframe
print(df.head())
```

```python
# prompt: Add a column isValidMobile of boolean which contains the values true or false against each value from phoneNumber column If a phone number is a valid Indian phone number.
# CREATE YOUR OWN LOGIC, DONT USE ANY EXTERNAL PYTHON LIBRARY FOR VALIDATION
# true : indicates number is valid
# false : indicates number is not valid
# Problem Statement: Design a logic to check whether a phone number is v

def is_valid_indian_number(phone_number):
  if not isinstance(phone_number, str):
    return False
  if len(phone_number) != 10:
    return False
  if not phone_number.isdigit():
    return False
  if not phone_number.startswith(('6', '7', '8', '9')):
    return False
  return True

# Apply the function to create the 'isValidMobile' column
df['isValidMobile'] = df['patientDetails_phoneNumber'].apply(is_valid_indian_number)

# Display the updated dataframe
print(df.head())
```
